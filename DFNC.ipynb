```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar datos omitiendo la fila de encabezado
df = pd.read_excel("dfncpapc.xlsx",
                  sheet_name="2. Precio Interno Mensual",
                  skiprows=5,  # Saltar una fila adicional para omitir los encabezados
                  header=None)

# 2. Seleccionar columnas correctas (ajustar índices según necesidad)
mes = df.iloc[:, 3]  # Columna de fechas (ajusta el índice)
precio = df.iloc[:, 4]  # Columna de precios (ajusta el índice)

# 3. Limpieza básica (convertir tipos y eliminar filas no válidas)
mes = pd.to_datetime(mes, errors='coerce')
precio = pd.to_numeric(precio, errors='coerce')
df_clean = pd.DataFrame({'Mes': mes, 'Precio': precio}).dropna()

# 4. Gráfico
plt.figure(figsize=(12, 6))
plt.plot(df_clean['Mes'], df_clean['Precio'], 'r-', linewidth=1)
plt.title("Evolución del Precio Interno del Café")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar los datos correctamente
file_path = "dfncpapc.xlsx"
df = pd.read_excel(file_path,
                  sheet_name="9. Producción mensual",
                  skiprows=7,  # Saltar las filas de encabezado
                  header=None,
                  usecols=[3, 4])


# 2. Asignar nombres a las columnas
df.columns = ['Mes', 'Produccion']


# 3. Limpieza de datos
df['Mes'] = pd.to_datetime(df['Mes'], errors='coerce')
df['Produccion'] = pd.to_numeric(df['Produccion'], errors='coerce')
df_clean = df.dropna()

# 4. Verificación
print("=== PRIMERAS FILAS ===")
print(df_clean.head())
print("\n=== ÚLTIMAS FILAS ===")
print(df_clean.tail())

# 5. Gráfico de producción
plt.figure(figsize=(12, 6))
plt.plot(df_clean['Mes'], df_clean['Produccion'], 'g-', linewidth=1.5)
plt.title("Producción Mensual de Café (Miles de sacos de 60 Kg)")
plt.xlabel("Mes")
plt.ylabel("Producción")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
