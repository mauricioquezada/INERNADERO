```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar datos omitiendo la fila de encabezado
df = pd.read_excel("dfncpapc.xlsx",
                  sheet_name="2. Precio Interno Mensual",
                  skiprows=5,  # Saltar una fila adicional para omitir los encabezados
                  header=None)

# 2. Seleccionar columnas correctas (ajustar índices según necesidad)
mes = df.iloc[:, 3]  # Columna de fechas (ajusta el índice)
precio = df.iloc[:, 4]  # Columna de precios (ajusta el índice)

# 3. Limpieza básica (convertir tipos y eliminar filas no válidas)
mes = pd.to_datetime(mes, errors='coerce')
precio = pd.to_numeric(precio, errors='coerce')
df_clean = pd.DataFrame({'Mes': mes, 'Precio': precio}).dropna()

# 4. Gráfico
plt.figure(figsize=(12, 6))
plt.plot(df_clean['Mes'], df_clean['Precio'], 'r-', linewidth=1)
plt.title("Evolución del Precio Interno del Café")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1️⃣ Verificar y cargar el archivo de datos
file_path = "dfncpapc.xlsx"

if file_path not in os.listdir():
    try:
        from google.colab import files
        print(" Sube el archivo 'dfncpapc.xlsx'.")
        uploaded = files.upload()  # Para uso en Google Colab
    except:
        print(f"❌ Error: No se encontró el archivo {file_path} en el directorio actual.")
        exit()

# 2️⃣ Cargar y preparar los datos
try:
    # Leer datos de Excel
    df = pd.read_excel(
        file_path,
        sheet_name="9. Producción mensual",
        skiprows=7,
        header=None,
        usecols=[3, 4]  # Columnas D (3) y E (4)
    )

    # Asignar nombres a las columnas
    df.columns = ['Mes', 'Produccion']

    # Limpieza de datos
    df['Mes'] = pd.to_datetime(df['Mes'], errors='coerce')
    df['Produccion'] = pd.to_numeric(df['Produccion'], errors='coerce')
    df_clean = df.dropna().reset_index(drop=True)

    # Verificación de datos
    print("=== RESUMEN DE DATOS ===")
    print(f"Período cubierto: {df_clean['Mes'].min().strftime('%Y-%m')} a {df_clean['Mes'].max().strftime('%Y-%m')}")
    print(f"Total de registros válidos: {len(df_clean)}")
    print("\n=== PRIMERAS FILAS ===")
    print(df_clean.head())
    print("\n=== ÚLTIMAS FILAS ===")
    print(df_clean.tail())

    # 3️⃣ Análisis gráfico
    plt.figure(figsize=(14, 8))

    # Gráfico 1: Serie temporal original
    plt.subplot(2, 1, 1)
    plt.plot(df_clean['Mes'], df_clean['Produccion'], 'g-', linewidth=1.5)
    plt.title("📈 Producción Mensual de Café (Miles de sacos de 60 Kg)")
    plt.xlabel("Mes")
    plt.ylabel("Producción")
    plt.grid(True)
    plt.xticks(rotation=45)

    # Gráfico 2: Serie con suavizado
    df_clean['Produccion_Suavizada'] = df_clean['Produccion'].rolling(window=12, center=True).mean()

    plt.subplot(2, 1, 2)
    plt.plot(df_clean['Mes'], df_clean['Produccion'], label="Producción Real", alpha=0.5)
    plt.plot(df_clean['Mes'], df_clean['Produccion_Suavizada'],
             label="Tendencia (Promedio Móvil 12 meses)", color='red', linewidth=2)
    plt.title("📊 Producción con Suavizado de Tendencia")
    plt.xlabel("Mes")
    plt.ylabel("Producción (Miles de sacos)")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

    # 4️ Análisis estacional (opcional)
    if len(df_clean) >= 24:  # Necesitamos al menos 2 años de datos
        df_clean['Año'] = df_clean['Mes'].dt.year
        df_clean['Mes_num'] = df_clean['Mes'].dt.month

        # Gráfico de cajas por mes para ver estacionalidad
        plt.figure(figsize=(12, 6))
        df_clean.boxplot(column='Produccion', by='Mes_num', grid=False)
        plt.title("Variación Estacional de la Producción")
        plt.suptitle("")  # Eliminar título automático
        plt.xlabel("Mes del año")
        plt.ylabel("Producción")
        plt.xticks(range(1,13), ['Ene','Feb','Mar','Abr','May','Jun',
                                'Jul','Ago','Sep','Oct','Nov','Dic'])
        plt.show()

except Exception as e:
    print(f"❌ Error durante el procesamiento: {str(e)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Verificar si el archivo está disponible
file_path = "dfncpapc.xlsx"

if file_path not in os.listdir():
    from google.colab import files
    print(" Sube el archivo 'dfncpapc.xlsx'.")
    uploaded = files.upload()  # Abre una ventana para subir el archivo

# 1. Cargar los datos correctamente
df = pd.read_excel(file_path,
                   sheet_name="9. Producción mensual",  # Nombre de la hoja con datos de exportación
                   skiprows=7,  # Saltar filas innecesarias
                   header=None,  # No hay encabezados definidos
                   usecols=[3, 4])  # Seleccionar las columnas necesarias

# 2. Asignar nombres a las columnas
df.columns = ['Mes', 'Exportación']

# 3. Limpieza de datos
df['Mes'] = pd.to_datetime(df['Mes'], errors='coerce')  # Convertir a formato fecha
df['Exportación'] = pd.to_numeric(df['Exportación'], errors='coerce')  # Convertir a número
df_clean = df.dropna()  # Eliminar filas con datos faltantes

# 📈 4. Graficar la exportación mensual de café
plt.figure(figsize=(14, 6))
plt.bar(df_clean['Mes'], df_clean['Exportación'], color='brown', alpha=0.6, edgecolor='black')

#  Estilo y etiquetas
plt.title("Exportación Mensual de Café", fontsize=16)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Miles de sacos de 60 Kg de café verde equivalente", fontsize=12)
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os



#  Verificar si el archivo está disponible
file_path = "dfncpapc.xlsx"

if file_path not in os.listdir():
    from google.colab import files
    print(" Sube el archivo 'dfncpapc.xlsx'.")
    uploaded = files.upload()

#  Cargar los datos
df = pd.read_excel(file_path,
                   sheet_name="9. Producción mensual",
                   skiprows=7,
                   header=None,
                   usecols=[3, 4])

# 🏷 Asignar nombres y limpiar datos
df.columns = ['Mes', 'Exportación']
df['Mes'] = pd.to_datetime(df['Mes'], errors='coerce')
df['Exportación'] = pd.to_numeric(df['Exportación'], errors='coerce')
df_clean = df.dropna()

# =============================================
# PARTE 2: Gráfico combinado
# =============================================

plt.figure(figsize=(16, 8))

# SUBGRÁFICO 1: Exportación mensual de café (original)
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, posición 1
plt.bar(df_clean['Mes'], df_clean['Exportación'], color='brown', alpha=0.6, edgecolor='black')
plt.title("Exportación Mensual de Café", fontsize=14)
plt.xlabel("Mes", fontsize=10)
plt.ylabel("Miles de sacos (60 Kg)", fontsize=10)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# SUBGRÁFICO 2: Datos de ejemplo (15-25 años)
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, posición 2

# Datos de ejemplo (simulando años de exportación)
años = list(range(12, 26))  # Del año 12 al 25
valores = [100, 120, 90, 150, 200, 180, 220, 250, 300, 280, 320, 340, 380, 400, 420, 450]

# Filtrar rango 15-25
años_filtrados = [año for año in años if 15 <= año <= 25]
valores_filtrados = [valores[i] for i in range(len(años)) if 15 <= años[i] <= 25]

# Gráfico de barras
plt.bar(años_filtrados, valores_filtrados, color='skyblue', edgecolor='black')
plt.title("Exportación Anual (Años 15-25)", fontsize=14)
plt.xlabel("Año", fontsize=10)
plt.ylabel("Miles de sacos (60 Kg)", fontsize=10)
plt.xticks(años_filtrados)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar layout
plt.tight_layout()
plt.show()
```
